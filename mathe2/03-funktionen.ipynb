### Plotting mit matplotlib

Dieser Abschnitt dient gleichzeitig als Einführung ins Plotten mit matplotlib. Hier sehen Sie ein Beispiel, wie die Funktion $f(x,a,b) = a\cdot x^2 + b$ geplottet wird. In matplotlib werden zum Plotten Punkte $x$ mit zugehörigen Funktionswerten $y$ benötigt. Es wird zunächst die Funktion definiert, danach werden Werte $x$ definiert, an denen die Funktion ausgewertet werden soll. Die Funktionswerte werden in $y$ gespeichert und dann wird $x$ gegen $y$ geplottet.

Führen Sie den Code aus, um die Grafik zu erzeugen.

```python
def f(x):
    """Definition der Funktion f(x,a,b)"""
    if x <= 3:
        return x**2 -4*x + 4
    else:
        return x*x - 8*x + 16

# Define values x: 100 values in the interval -2 to 2
x = linspace(0, 4, 100)

# Compute function values y for a=2 and b=-3. Note how python can automatically compute the function value for the complete array x.
y = zeros(100)
for i in range(100):
    y[i] = f(x[i])

# Create a new figure, label the axes and plot (x,y)
figure()
plot(x, y)

# Optional: title and axes labels
#title('$2 x^2 - 3$')
xlabel('x')
ylabel('y')
```
